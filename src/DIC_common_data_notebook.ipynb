### Identifying highly skewed columns
<i>Done by Abhinav Tembulkar</i><br>
We identify highly skewed columns and drop them from our dataset. 

Below we demonstrate how Currency column is highly skewed. We cannot use such columns to derive any meaningful hypothesis in our EDA

```python
df = pd.read_csv('../data/merged_dataset.csv')
df['currency'].value_counts()
```

```python
# Define a function to calculate skewness and drop columns based on a threshold

skewness_threshold=2
# Calculate skewness for each numerical column
skewness = df.select_dtypes(include=['number']).skew().abs()

# Display skewness values
print("Skewness of each numerical column:")
print(skewness)
```

```python
# Identify columns with skewness greater than the threshold
highly_skewed = skewness[skewness > skewness_threshold].index

# Display highly skewed columns
print("Highly skewed columns to be dropped:")
print(highly_skewed)

with open('../preprocessing/highly_skewed_columns.txt', 'w') as f:
    for cols in highly_skewed:
        f.write(cols + "\n")

# Drop these columns from the DataFrame
df.drop(columns=highly_skewed, inplace=True)

# Display the DataFrame after dropping highly skewed columns
print("DataFrame after dropping highly skewed columns:")
print(len(df.columns))
```

### Imputing missing values in dataset
<i>Done by Atharva Prabhu</i><br>
We use K-Nearest Neighbours to infer missing values in our dataset to make it suitable for EDA

```python
df = pd.read_csv('../data/merged_dataset_atul.csv')
```

```python
# df_cleaned = df.dropna(axis='columns', how='all')
df_cleaned = df.copy()
```

```python
detect_write_empty_columns(df_cleaned)
```

```python
df_cleaned.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import KNNImputer
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
```

```python
def detect_and_ordinal_encode(df):
    # Create a copy of the original dataframe
    df_encoded = df.copy()
    
    # Initialize a dictionary to store label encoders for each column
    label_encoders = {}

    # Loop through columns
    for col in df_encoded.columns:
        # Check if column is categorical (object type)
        if df_encoded[col].dtype == 'object':
            # Apply Label Encoding
            le = LabelEncoder()
            non_na_mask = df_encoded[col].notna()  # Mask for non-NaN values
            df_encoded.loc[non_na_mask, col] = le.fit_transform(df_encoded.loc[non_na_mask, col].astype(str))
            # df_encoded[col] = df_encoded[col].astype(str).apply(lambda x: le.fit_transform([x]) if pd.notna(x) else x)

            
            # Store the LabelEncoder for future use (e.g., for decoding or applying to test data)
            label_encoders[col] = le

    return df_encoded, label_encoders

encoded_df , labels_encoders = detect_and_ordinal_encode(df_cleaned)
```

```python
len(encoded_df.columns)
```

```python
encoded_df['Age'].value_counts()
```

```python
print(encoded_df.shape)
print(len(encoded_df.columns))
```

```python
encoded_df.head()
```

```python
knn_imputer = KNNImputer(n_neighbors=5)

# Perform imputation and convert the result back to a DataFrame
df_imputed = pd.DataFrame(knn_imputer.fit_transform(encoded_df), columns=encoded_df.columns)

# Ensure the encoded categorical columns are of integer type after imputation (since KNN Imputer returns floats)
df_imputed = df_imputed.apply(lambda x: x.astype(int) if x.name in labels_encoders else x)
```

<b>Decoding the encoded dataframe</b>

```python
df_imputed = pd.read_csv('../data/imputed_dataset.csv')
```

```python
def dataframe_decoder(df, labelEncoder):
    # Decode the values in df_encoded
    df_decoded = df.copy()

    for col, le in labelEncoder.items():
        # Apply the inverse transform to each encoded column
        df_decoded[col] = le.inverse_transform(df_decoded[col])

    return df_decoded
```

```python
df_decoded = dataframe_decoder(df_imputed, labels_encoders)
df_decoded.head()
```
